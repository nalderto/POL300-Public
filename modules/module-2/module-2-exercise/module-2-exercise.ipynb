## Relational Operators

We use relational operators in order to compare various values.

| Operator | Meaning |
| :--- | :--- |
| `==` | Equals |
| `!=` | Not equal |
| `>` | Greater than |
| `>=` | Greater than or equal to |
| `<` | Less than |
| `<=` | Less than or equal to |

It is important to keep in mind that relational operators simply return Boolean `True` or `False`. For instance: `print(5 > 1)` would print Boolean `True`.

### Examples

```python
print(1 == 1) # Prints True
print(1 != 1) # Prints False
print(1 > 3) # Print False
print(1 <= 1) # Prints True
print(1 < 7) # Prints True
```

## Boolean Operators

Sometimes it is useful to be able to check multiple variables when using conditionals. We can achieve that using _boolean operators_. These operators allow us to combine the expressions that are made using the relational operators above.

| Operator | When does it return `True` |
| :--- | :--- |
| `and` | Expressions on both sides of the `and` must be `True` |
| `or` | Only one side needs to be `True` |
| `not` | When the expression to its right is `False` |

### Examples

```python
print(True and True) # Prints True
print(True and False) # Prints False
print(False and False) # Print False
print(7 > 5 and 7 > 4) # Prints True

print(True or False) # Prints True
print(False or False) # Prints False
print(5 > 1 and 7 > 4) # Prints True

print(not False) # Prints True
print(not 7 > 4) # Prints False
```

## Conditionals

It is often extremely useful to be able to run a snippet of code depending on what a value of a variable is. Imagine you are creating a function to determine if a person can purchase alcohol in the United States. The function would look something like this:

```python
def canBuyAlcohol(age):
    if age >= 21:
        print("You can buy alcohol")
    elif age == 20:
        print("Wait until you are 21")
    else:
        print("Sorry, you are too young")
```

Let's unpack the different components of this code block.

### Clauses

There are three clauses in conditionals.

1. `if`
2. `elif`
3. `else`

The `if` statement is the only **required** clause in a conditional statement, as it starts every new conditional block. You cannot have `elif` or `else` clauses without first starting with an `if`. An expression follows the word `if`. If this expression is Boolean `True`, then the code following the if statement is ran. Keep in mind that indentation is very important. The code following a clause must be indented. The `elif` stands for "else if". Just like an `if` clause, it has an expression that follows the clause. Not all conditionals need an `elif` section. It is important to keep in mind that only a single code section will run, even if multiple `if` and `elif` clauses are Boolean `True`. For example:

```python
x = 3
if x > 2:
    print("This will print")
elif x > 1:
    print("This will NOT print")
```

However if we use two if statements \(which represent two separate conditionals\),

```python
x = 3
if x > 2:
    print("This will print")
if x > 1:
    print("This will print")
```

The `else` clause is the final section of any conditional. Like the `elif` section, it is completely optional. Unlike `if` and `elif` clauses, there is no expression following the clause. The `else` section is a "catch-all". If the none of the previous `if` or `elif` expressions were Boolean `True`, then the `else` code will run.

## Data Structures

There are several data structures that are built into Python, and even more that are available via external Python modules. We are going to focus on two for the time being: lists and dictionaries.

### Lists

Lists are exactly as they sound, they are just lists of data. The data that is contained within a lists can be of any type. You can have a string, float, integer, and Boolean all in the same list.

#### Declare Lists

```python
empty = []
presidents = ["Washington", "Adams", "Jefferson"]
stuff = ["Purdue", 3, 4.2, True]
```

Lists are created like any other variable, with the name of the list being followed by an equal sign. Lists are denoted by a pair of square brackets \(`[]`\) at the beginning and end. Each of the items in the list are separated by commas. The `presidents` list only contains strings, however you can mix variables in lists, as is seen with the `stuff` list.  The `empty` variable is still a list, however it is simply empty.  Since there are no values inside the square brackets.

#### Accessing Lists

```python
presidents = ["Washington", "Adams", "Jefferson"]
first = presidents[0] # Washington
```

In order to access one of the items in a list, you type the name of the list, followed by the respective index number surrounded by square brackets. Python uses _zero-based numbering_. This means that the first element in a list is accessed by using the index `0`. With the `presidents` list example above, in order to access each of the strings, you would use the following indices.

0. Washington
1. Adams
2. Jefferson

#### Accessing Range of Items

```python
presidents = ["Washington", "Adams", "Jefferson"]
twoThree = presidents[1:3] # Adams and Jefferson
```

The process for accessing multiple items in a list is similar to accessing a single item. Instead of only specifying a single index, you provide a range, divided by a colon \(`[start:end]`\). Keep in mind that the range starts with the first index and the ends with the second index \(non-inclusive\). In the example above, we have to do a range `1:3`, because we want to start with `"Adams"` \(index 1\) and end with `"Jefferson"` \(index 2\). Remember, we start counting with 0 in Python.  We put `3` since we want to stop right before index `3` \(which doesn't exist in this case, since the list ends\).

#### Add Items to a List

```python
presidents = ["Washington", "Adams", "Jefferson"]
presidents.append("Monroe") # ["Washington", "Adams", "Jefferson", "Monroe"]
presidents.insert(3, "Madison") # ["Washington", "Adams", "Jefferson", "Madison", "Monroe"]
presidents.extend(["Adams", "Jackson", "Van Buren"]) # ["Washington", "Adams", "Jefferson", "Madison", "Monroe", "Adams", "Jackson", "Van Buren"]
```

The `list.append(item)` function will add the specified value to the end of the list. You type the name of your respective list \(like `presidents`\), then a `.append()`, with item you would like to add to the list in the parentheses.  It is important to note that only one value can be added at a time with `.append()`.

The `list.insert(index, item)` function will add the passed in item to the specified index. The preexisting items in the list following the insertion location will be shifted over by a single index \(i.e. `"Monroe"` goes from index 3 to index 4\).

The `list.extend(different_list)` allows you to add multiple new elements to a pre-exisiting list.  This function takes another, different list as a parameter.  It will take all the elements from the different list and add them to the list where `.extend()` is called from.  

#### Remove Items from a List

```python
states = ["Alaska", "California", "Indiana", "Washington, D.C.", "Wyoming"]
states.remove("Washington, D.C.")
```

Removing an item from a list is as simple as using the `list.remove(item)` notation, with the value of the item between the parentheses.

### Dictionaries

Dictionaries are somewhat similar similar to lists with some key differences. Dictionaries are called "dictionaries" as they act very similar to how you would look up a definition. In a book dictionary, you look up a definition by searching by the word in question. If we were to think of a dictionary as a data structure, our items would be the word and definition pair. In Python we have key and value pairs. You can think of the key being the word and value being the definition. You would use the key to then find the definition. Here is an example:

```python
purdue = {"founded": 1869, "president": "Mitch Daniels", "conference": "Big Ten", "city": "West Lafayette"}
```

In this example, `purdue` is the name of the dictionary. The keys are `"founded", "president", and "conference", and "city"` and the corresponding values are `1869, "Mitch Daniels", "Big Ten", and "West Lafayette"`.

#### Accessing Dictionary Item

```python
purdue = {"founded": 1869, "president": "Mitch Daniels", "conference": "Big Ten", "city": "West Lafayette"}
whereIsPurdue = purdue["city"] # West Lafayette
```

Accessing a value is a matter of typing in the name of the dictionary, followed by square brackets containing the key name \(with the quotation marks\).

#### Adding Dictionary Items

```python
purdue = {"founded": 1869, "president": "Mitch Daniels", "conference": "Big Ten", "city": "West Lafayette"}
purdue["state"] = "Indiana"
```

To add an item to a preexisting dictionary, you type the name of the dictionary, followed by square brackets with the new key name. You then set this equal to the value of that key. From that point forward, the dictionary with contain that key-value pair \(unless you delete it\).

#### Editing Dictionary Items
```
purdue = {"founded": 1869, "president": "Mitch Daniels", "conference": "Big Twelve", "city": "West Lafayette"}
purdue["conference"] = "Big Ten"
```
To change the value of dictionary key, simply access the dictionary key using the square bracket notation (`purdue["conference"]`), then change the value by using an equal sign (`= "Big Ten"`).  It's very similar to changing the value of a regular variable, except you first have to access the appropriate dictionary entry.  

#### Deleting Dictionary Items

```python
purdue = {"founded": 1869, "president": "Mitch Daniels", "conference": "Big Ten", "city": "West Lafayette"}
purdue.pop("city") # purdue no longer includes the "city" key
```

The `dictionary.pop(key)` notation is used remove a key-value pair from a dictionary.

## Loops

Frequently, we want to run the same piece of code over and over again. For example, maybe we want to iterate through a data structure \(like a list\), or perhaps we want to continue running a piece of code until a condition is met. It is during these circumstance where we use _loops_. There are two types of loops in Python `for` and `while`.

### `for` Loop

A `for` loop is used when the number of times we want to run a piece of code is predetermined by either the programmer or by an existing variable. This is what a `for` loop looks like:

```python
for i in range(4):
    print(i)
```

Let's break down what this code snippet does.

* Every `for` loop must start with the keyword `for`
* The `i` simply represents one of the values in the data that we are iterating through.  It doesn't have to be `i` \(although that is commonly used by programmers\).  Instead we could use `number`.
* `in` is another keyword that is always used with `for` loops.  It essentially tells Python what we are looping through
* `range(4)` is a simple function that expands into `[0,1,2,3]`.  It creates a list starting a 0 and ending at the specified number \(non-inclusive\).  We'll expand upon this in a little bit.
* The header of `for` loops end with a colon \(`:`\).  This creates a dividing point between the loop header and the loop body.  
* `print(i)` will print the value of `i` which was declared in the loop header.  Recall that the loop variable could be named nearly anything.  If the variable was named `number`, then we would have used `print(number)`.  As the loop progresses, the value of `i` changes to each of the corresponding values in the data structure specified \(`range(4)` in this instance\). **Please note that the loop body is indented from the loop header, this is required**

#### `range()` Function

The range is quite useful with `for` loops as they allow you to count. This function takes on three forms:

1. `range(stop)` Creates a list from 0 to `stop` \(non-inclusive\) and counts by one.
2. `range(start, stop)` Creates a list from `start` to `stop` \(non-inclusive\) and counts by one.
3. `range(start, stop, interval)` Creates a list from `start` to `stop` \(non-inclusive\) and counts by `interval`.  

For example `range(2, 8, 2)` would create a list `[2, 4, 6]`.

#### Looping through Lists
You can also use `for` loops to loop through lists as well.

```python
items = ["file", "steam", "water", "light"]

for item in items:
    print(item)
```

This code segment will print each of words in the `items` list in order.  Essentially, the value of the variable `item` will change to each of the values in the list with each iteration of the loop.  Code that is within the scope of the loop will be able to use the `item` variable.  Keep in mind that the variable doesn't have to be called `item`.  It could be called anything, like `word` for instance.  

### A Note on the Importance of Indentation
Run the following code segment:

````python
def function_1():
    for i in range(10):
        if i == 4:
            print('We made it to 4!')
        return i
    
def function_2():
    for i in range(10):
        if i == 4:
            print('We made it to 4!')
            return i

print('This is function_1():')     
print(function_1())
print("")
print('This is function_2():')  
print(function_2())
````

As you can see, the functions one and two contain the exact same text, however the indentation is different.  The `return i` line is under the scope of the `for` statement in `function_1()`, while the `return` statement is scoped under the `if` statement in `function_2()`.  

In `function_1()`, the first iteration of the loop results in the function returning `0`.  This is because in the first iteration of the loop, where `i = 0`, we run `return i`, since it is in the indentation scope of the `for` loop.  

However, in `function_2()`, the function waits until `i == 4` for it to return `i`. This is because the `return i` line is indented in the purview of `if i == 4:`.  Python only runs the code within the `if i == 4:` scope when `i == 4`.

The example is meant to show how important indentation is to your code.  You need to pay close attention to ensure you get the results that you expect.

### `while` Loop

`while` loops are typically used when the stopping point for a loop isn't necessarily known by the programmer ahead of time. For example, if we are processing data from an Excel spreadsheet, we may not know how many rows will be in that file. We don't want to have to manually check each time and change our code accordingly. Instead we could simply use a while loop that will run until no more data is left in the spreadsheet. While loops like like this:

```python
x = 1
while x < 4:
    print(x)
    x += 1
```

The example above is not a typical use case for `while`. Frankly, it would be better to use a `for` loop in this case, however this is simply a demonstration.

* Similar to `for` loops, `while` loops must also start with the `while` keyword.
* We next put our end condition for the loop.  In order for the body of the loop to run, this condition must be Boolean `True`.  If we put `while True:` the loop would typically never end, which is known as an _infinite loop_.  **You need to be careful about potentially creating infinite loops, as they can potentially cause your computer or Jupyter to crash.**  Since we initially set `x=1`, and `1 < 4` is `True`, the loop starts.  
* `print(x)` and `x += 1` are the loop body.  **Notice that the loop body is indented.**  We do `x += 1` to increment the value of `x` to ensure that the loop eventually terminates.

### `break` Statement

Sometimes it is useful to terminate a loop prematurely. The `break` statement is used to achieve this. Consider the following example:

```python
for i in range(5):
    if i == 3:
        break
    print(i)
```

This loop would print:

```python
0
1
2
```

This is just a standard `for` loop that will print the values of `i`. However, once `i == 3`, the loop will terminate, and your program will continue executing the code that follows the loop.

### `continue` Statement

`continue` is a little different compared to `break`. Instead of terminating the whole loop, it just ends that iteration. Consider the following example:

```python
for i in range(5):
    if i == 3:
        continue
    print(i)
```

This loop would print:

```python
0
1
2
4
```

Notice how the loop continues after reaching the `continue`. All the `continue` did in this instance is skip the `print(i)` statement when `i == 3`.
